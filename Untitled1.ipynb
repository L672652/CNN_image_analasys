```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import os
import matplotlib.pyplot as plt
from keras.utils.np_utils import to_categorical
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from sklearn.metrics import confusion_matrix
from tensorflow.keras.metrics import Precision, BinaryAccuracy
data = tf.keras.utils.image_dataset_from_directory('EuroSAT_RGB/EuroSAT_RGB', image_size=(64,64))
data_iterator = data.as_numpy_iterator()
batch = data_iterator.next()
scaled = batch[0]/255
scaled.max()

data = data.map(lambda x,y: (x/255, tf.one_hot(y, depth=10)))
scaled_iterator = data.as_numpy_iterator()
batch = scaled_iterator.next()

batch[0].min()

len(data)

train_size = int(len(data)*.7)
val_size = int(len(data)*.2)
test_size = int(len(data)*.1)

train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)

X_val_list = []
Y_val_list = []

model = Sequential()
model.add(Conv2D(16,(3,3),1,activation='relu',input_shape=(64,64,3)))
model.add(MaxPooling2D())

model.add(Conv2D(32,(3,3),1,activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(16,(3,3),1,activation='relu'))
model.add(MaxPooling2D())

model.add(Flatten())


model.add(Dense(256,activation='relu'))
model.add(Dense(10, activation = "softmax"))


model.compile('adam' , loss = "categorical_crossentropy", metrics=["accuracy"])
model.summary()

logdir='logs'
print(train.element_spec)
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
hist = model.fit(train,epochs=20,validation_data=val, callbacks=[tensorboard_callback])

model.save('SateliteModel')

acc = BinaryAccuracy()

for batch in test.as_numpy_iterator():
    X,Y = batch
    yhat = model.predict(X)
    acc.update_state(Y,yhat)

print(f'Treffsikkerhet:{acc.result().numpy()}')

for features, labels in val:
    X_val_list.append(features)
    Y_val_list.append(labels)
    
X_val = tf.concat(X_val_list, axis=0)
Y_val = tf.concat(Y_val_list, axis=0)

plt.plot(hist.history['accuracy'], label='accuracy')
plt.plot(hist.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoker')
plt.ylabel('Treffsikkehet')
plt.legend(loc='lower right')

test_loss, test_acc = model.evaluate(test, verbose=2)

Y_pred = model.predict(X_val)
Y_pred_classes = np.argmax(Y_pred,axis = 1) 
Y_true = np.argmax(Y_val,axis = 1) 
confusion_mtx = confusion_matrix(Y_true, Y_pred_classes) 
f,ax = plt.subplots(figsize=(10, 10))
sns.heatmap(confusion_mtx, annot=True, linewidths=0.02,cmap="Blues",linecolor="gray", fmt= '.1f',ax=ax)
plt.xlabel("Predikert Terreng")
plt.ylabel("Sanne Terreng")
plt.title("Matrise som viser treffsikkerhet (confusion matrix)")
plt.show()
 
 
```
